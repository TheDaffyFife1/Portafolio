```python
cimport matplotlib.pyplot as plt

# Datos
resoluciones = [(3036, 4048), (1518, 2024), (1138, 1517), (759, 1012), (379, 505), (189, 252)]
t1 = [
    [383, 385, 380, 440, 379, 381, 381, 386, 382, 389],
    [164, 159, 164, 205, 166, 173, 161, 163, 163, 166],
    [92, 89, 97, 93, 90, 92, 93, 94, 108, 94],
    [52, 45, 48, 46, 46, 46, 45, 44, 51, 47],
    [23, 22, 31, 21, 26, 21, 23, 23, 23, 24],
    [13, 12, 12, 13, 11, 13, 12, 12, 11, 16]
]
t6 = [
    [239, 231, 237, 249, 245, 235, 250, 243, 231, 235],
    [81, 78, 85, 81, 87, 84, 79, 78, 89, 78],
    [62, 54, 72, 58, 58, 57, 56, 55, 59, 62],
    [44, 38, 42, 37, 43, 41, 39, 40, 40, 38],
    [27, 30, 30, 25, 27, 25, 29, 28, 31, 27],
    [16, 16, 15, 15, 15, 15, 15, 16, 14, 16]
]

# Calcular la aceleración en función de t1 y t6 para ambas resoluciones
aceleraciones = []
for i in range(len(resoluciones)):
    aceleraciones_resolucion = [((t1[i][j] - t6[i][j]) / t1[i][j]) * 100 for j in range(len(t1[i]))]
    aceleraciones.append(aceleraciones_resolucion)

# Calcular el promedio de la aceleración para cada resolución
promedios_aceleracion = [sum(aceleraciones[i]) / len(aceleraciones[i]) for i in range(len(aceleraciones))]

# Crear la gráfica de promedios de aceleración
plt.figure(figsize=(10, 6))  # Establecer el tamaño de la figura
plt.bar([f"{res[0]}x{res[1]}" for res in resoluciones], promedios_aceleracion, color='blue')

# Etiquetas de los ejes
plt.xlabel('Resolución')
plt.ylabel('Promedio de Aceleración (%)')

# Título del gráfico
plt.title('Promedio de Aceleración vs. Resolución')

# Mostrar el gráfico
plt.xticks(rotation=45)  # Rotar etiquetas en el eje x para una mejor legibilidad
plt.tight_layout()  # Ajustar el diseño
plt.grid(True)
plt.show()

# Identificar el punto en el que la aceleración deja de mejorar significativamente
for i in range(len(resoluciones)):
    print(f"Resolución {resoluciones[i]} - Promedio de Aceleración: {promedios_aceleracion[i]:.2f}%")
```

La gráfica muestra la relación entre la resolución y el promedio de aceleración para varios conjuntos de datos. Se pueden observar varias tendencias interesantes:

Variación en los promedios de aceleración: Los promedios de aceleración varían significativamente en función de la resolución. Algunos conjuntos de datos tienen promedios de aceleración positivos, mientras que otros tienen valores negativos.

Resoluciones más altas tienden a tener promedios de aceleración positivos: En general, parece haber una tendencia hacia promedios de aceleración positivos a medida que aumenta la resolución. Esto sugiere que en resoluciones más altas, la aceleración tiende a ser mayor.

Excepciones: Sin embargo, también hay excepciones a esta tendencia general. Por ejemplo, el conjunto de datos con resolución más baja (189) tiene un promedio de aceleración negativo, lo que indica que no siempre se cumple la relación directa entre resolución y aceleración.
