```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#2021

```python
df_2021 = pd.read_excel('/content/Cantidad de compras por producto 2021.xlsx')
```

```python
df_2021.head()
```

```python
col = ['Material']
df_2021 = df_2021.drop(df_2021.index[-1])
PrimT = df_2021.loc[:, col]
SegT  = df_2021.loc[:, col]
TerT  = df_2021.loc[:, col]
CuarT = df_2021.loc[:, col]
```

```python
#Generacion de la separacion por trimestre
Fptrim= ['Enero','Febrero','Marzo']
Sptrim= ['Abril', 'Mayo', 'Junio']
Tptrim= ['Julio', 'Agosto', 'Septiembre']
Cptrim= ['Octubre', 'Noviembre', 'Diciembre']
```

```python
#Trimestral = pd.DataFrame()
PrimT['Primer trimestre'] = df_2021[Fptrim].sum(axis=1)
SegT['Segundo trimestre'] = df_2021[Sptrim].sum(axis=1)
TerT['Tercer trimestre'] = df_2021[Tptrim].sum(axis=1)
CuarT['Cuarto trimestre'] = df_2021[Cptrim].sum(axis=1)
```

```python
PrimT = PrimT.sort_values(by='Primer trimestre', ascending=False)
productos1_2021 = PrimT.head(5)

SegT = SegT.sort_values(by='Segundo trimestre', ascending=False)
productos2_2021 = SegT.head(5)

TerT = TerT.sort_values(by='Tercer trimestre', ascending=False)
productos3_2021 = TerT.head(5)

CuarT = CuarT.sort_values(by='Cuarto trimestre', ascending=False)
productos4_2021 = CuarT.head(5)
```

```python
plt.figure(figsize=(10, 6))
ax = plt.subplot()

productos_plt_2021 = productos1_2021['Material']
ventas_2021 = productos1_2021['Primer trimestre']

plt.barh(productos_plt_2021, ventas_2021, color='skyblue')

plt.xlabel('Ventas Totales')
plt.ylabel('Productos')
plt.title('Top 5 Productos Más Vendidos en el primer trimestre de 2021')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = plt.subplot()

t1productos_plt_2021 = productos2_2021['Material']
ventas_2021 = productos2_2021['Segundo trimestre']

plt.barh(productos_plt_2021, ventas_2021, color='skyblue')

plt.xlabel('Ventas Totales')
plt.ylabel('Productos')
plt.title('Top 5 Productos Más Vendidos en el segundo trimestre de 2021')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = plt.subplot()

productos_plt_2021 = productos3_2021['Material']
ventas_2021 = productos3_2021['Tercer trimestre']

plt.barh(productos_plt_2021, ventas_2021, color='skyblue')

plt.xlabel('Ventas Totales')
plt.ylabel('Productos')
plt.title('Top 5 Productos Más Vendidos en el tercer trimestre de 2021')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = plt.subplot()

productos_plt_2021 = productos4_2021['Material']
ventas_2021 = productos4_2021['Cuarto trimestre']

plt.barh(productos_plt_2021, ventas_2021, color='skyblue')

plt.xlabel('Ventas Totales')
plt.ylabel('Productos')
plt.title('Top 5 Productos Más Vendidos en el cuarto trimestre de 2021')

plt.show()
```
